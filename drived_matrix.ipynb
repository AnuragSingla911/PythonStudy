```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
odi = pd.read_csv('https://raw.githubusercontent.com/AnuragSingla911/PythonStudy/master/odi.csv')
```

```python
odi.head(10)
```

```python
odi['century'] = odi.Runs >= 100
```

```python
odi.head()
```

```python
odi = odi[odi.Country == 'India']
odi = odi[odi.century == True]
```

```python
odi['Date'] = pd.to_datetime(odi['MatchDate'])
```

```python

odi['year'] = odi.Date.dt.year
odi.head()
```

```python
gb = odi.groupby(by = 'year')['century']
gb.describe().sort_values('count')
```

```python
grades = pd.read_csv('grades.csv')
```

```python
grades.head()
```

```python
grades['isZip_submission'] = grades['submission'].str.endswith('.zip')
```

```python
total_rows = grades.shape[0]
zip_submission = grades[grades['isZip_submission'] == True].shape[0]
percentage = zip_submission/total_rows*100
print(percentage)
```

```python
grades['Date'] = pd.to_datetime(grades['submit_time'])
```

```python
target_datetime = pd.to_datetime('2017-01-03 23:59:59')
```

```python
grades['on_time_submission'] = grades.Date <= target_datetime
```

```python
grades.head()
```

```python
grades['Exact_Date'] = grades.Date.dt.date
```

```python
grades.head()
```

```python
gb = grades.groupby('Exact_Date')
gb.count()
```

```python
grades['Exact_Hour'] = grades.Date.dt.hour
```

```python
grades.head()
```

```python
gb = grades.groupby('Exact_Hour')
gb.count()
```

```python
gb = grades['Exact_Hour'].value_counts()
gb.plot(kind = 'hist', bins = 24)
```
