```python
import seaborn as sns
df = pd.read_csv('currencies.csv')
df.columns
```

```python

df = df[['Euro','Japanese Yen','U.K. Pound Sterling','U.S. Dollar','Chinese Yuan','Australian Dollar','Indian Rupee']]
correlation_matrix = df.corr()
```

```python
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=10)
plt.title('Correlation Matrix')
plt.show()
```

```python
df = pd.read_csv('nas.csv')
df.columns
```

```python
df1 = df[['Mother.edu', 'Siblings']]
df1.Siblings.value_counts()

category_to_numeric = {
    '2 siblings': 2,
    '1 sibling': 1,
    '3 siblings': 3,
    '4+ siblings': 4,
    'Single child': 0  # You can assign any numeric value to 'Single child'
}

df1['Numeric_Siblings'] = df1['Siblings'].map(category_to_numeric)
```

```python
df1 = df1[['Mother.edu','Numeric_Siblings']]
```

```python
df1.head()
```

```python
gb = df1.groupby('Mother.edu')
gb.describe()
```
